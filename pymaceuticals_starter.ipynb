```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"
```

```python
# Read the Mouse and Drug Data and the Clinical Trial Data
# using pandas to read csv file
mouse_df = pd.read_csv(mouse_drug_data_to_load)
clinical_df = pd.read_csv(clinical_trial_data_to_load)

#Checking data to make sure all elements are present as well as to identify the best way to merge both datasets
#mouse_df.head()
clinical_df.head()
```

```python
# Combining the data using groupby Mouse ID
data_group = clinical_df.merge(mouse_df, on = 'Mouse ID')

# Display the data table for preview
data_group.head()
#data_group.tail()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
mean_tumor = data_group.groupby(["Drug", "Timepoint"]).mean()["Tumor Volume (mm3)"] 
# comm: using groupby on the previously grouped data
mean_tumor
```

```python
# Convert to DataFrame and preview
mean_tumor_df = pd.DataFrame(mean_tumor)
mean_tumor_df
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint

vol_tumor = data_group.groupby(["Drug", "Timepoint"]).sem()["Tumor Volume (mm3)"]

# Convert to DataFrame
vol_tumor = pd.DataFrame(vol_tumor) #repeating the same step for mean_tumor

# Preview DataFrame
vol_tumor.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
#Reset the index of the DataFrame, and use the default one instead.

mean_tumor_def = mean_tumor.reset_index()
vol_tumor_def = vol_tumor.reset_index()


#Using pivot to return reshaped dataFrame organized by given index / column values.
mean_tumor_format = mean_tumor_def.pivot(index="Timepoint", 
                              columns="Drug")["Tumor Volume (mm3)"]

vol_tumor_format = vol_tumor_def.pivot(index="Timepoint", 
                              columns="Drug")["Tumor Volume (mm3)"]

# Preview that Reformatting worked

mean_tumor_format.head()
#vol_tumor_format.head()
```

```python
# Generate the Plot (with Error Bars)


plt.errorbar(vol_tumor_format.index, vol_tumor_format.index["Capomulin"], yerr=mean_tumor_format["Capomulin"],color="r", marker="o", markersize=5, linestyle="dashed", linewidth=0.50)

plt.errorbar(vol_tumor_format.index, vol_tumor_format["Infubinol"], yerr=mean_tumor_format["Infubinol"], color="b", marker="^", markersize=5, linestyle="dashed", linewidth=0.50)

plt.errorbar(vol_tumor_format.index, vol_tumor_format["Ketapril"], yerr=mean_tumor_format["Ketapril"], color="g", marker="s", markersize=5, linestyle="dashed", linewidth=0.50)

plt.errorbar(vol_tumor_format.index, vol_tumor_format["Placebo"], yerr=mean_tumor_format["Placebo"], color="k", marker="d", markersize=5, linestyle="dashed", linewidth=0.50)

plt.title("Tumor Response to Treatment", fontType = 'Helvetica', FontSize = 14)
plt.ylabel("Tumor Volume (mm3)")
plt.xlabel("Time (Days)")
plt.grid(True)
plt.legend(loc="best", fontsize="small", fancybox=True)

# Show the Figure
plt.show()

# Save the Figure
plt.savefig("Fig1.png")
```
